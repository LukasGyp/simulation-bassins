```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as anim
```

```python
df = pd.read_csv("data.csv", header=None, names=['t', 'h1', 'h2', 'h3', 'h4'])
df
```

```python
t = df['t'].values
h1 = df['h1'].values
h2 = df['h2'].values
h3 = df['h3'].values
h4 = df['h4'].values
```

```python
plt.plot(t, h1, label='h1')
plt.plot(t, h2, label='h2')
plt.plot(t, h3, label='h3')
plt.plot(t, h4, label='h4')
plt.xlabel('time [h]')
plt.ylabel('height [m]')
plt.legend()
plt.ylim(0, 4)
plt.show()
```

```python
# Create an animation (default: 0~500h, modifiable by interval argment of FuncAnimation())
fig = plt.figure(figsize = (10, 6))

def update(i):
    plt.cla()
    plt.bar([1, 2, 3, 4], [h1[i], h2[i], h3[i], h4[i]])
    plt.title(f'{t[i]} h')
    plt.ylim(0, 4)

ani = anim.FuncAnimation(fig, update, interval = 500, frames = 500)

ani.save("simulation.gif", writer = 'pillow')
```
